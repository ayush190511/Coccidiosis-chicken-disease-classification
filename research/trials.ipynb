### 2. In the same file, we are using "@ensure_annotations", why?

```python
def getProduct(x:int, y:int) -> int:
    return x * y
```

```python
getProduct(3, 4)
```

```python
getProduct(x=3, y="4")
```

This is not something we wanted...  
So we are using @ensure_annotation

```python
from ensure import ensure_annotations
```

```python
@ensure_annotations

def getProduct(x:int, y:int) -> int:
    return x * y
```

```python
getProduct(3, 4)
```

```python
getProduct(x=3, y="4")
```

As you can see that now we are getting the error of wrong multiplication.  
This "@ensure_annotation" ensures that no wrong arguments are passthrough the function.
